Each node needs to have a position stored as an x,y point for drawing with matplotlib. I think I'll store those in a second dictionary that represents the positions for each node. Here is the graph I'll be working with to start:



```python
import matplotlib.pyplot as plt
from matplotlib import collections as mc
```

```python
graph = {'a':{'b','d'},
         'b':{'a','e','f'},
         'c':{'d','e'},
         'd':{'c','e'},
         'e':{'b','c','d','f'},
         'f':{'b','e'}
        }

positions = {'a':(1,1),
             'b':(5,3),
             'c':(3,4),
             'd':(2,9),
             'e':(7,8),
             'f':(9,6)
            }
```

```python
def generate_edges(graph):
    edges = []
    for node in graph:
        for neighbor in graph[node]:
            if {node, neighbor} not in edges:
                edges.append({node, neighbor})
    return edges
```

```python
generate_edges(graph)
```

```python
def draw_graph(graph, positions, show_edges=True):
    #create the labels
    plt.xlabel('x axis')
    plt.ylabel('y axis')
    plt.title('Graph Nodes')
    #define the size
    fig = plt.gcf()
    fig.set_size_inches(9, 8)
    fig.savefig('test2png.png', dpi=100)

    #plot the edges if desired
    if show_edges:
        edge_list = generate_edges(graph)
        for edge in edge_list:
            x, y = [], []
            for point in edge:
                x.append(positions[point][0])
                y.append(positions[point][1])
                plt.plot(x, y, color = 'g')
    
    #plot the nodes
    for node in positions:
        x, y = positions[node]
        plt.scatter(x, y, label = "NODES", color = 'blue', marker = f'${node.upper()}$', s=150)

    #plt.legend()
    plt.show()
```

```python
draw_graph(graph, positions)
```
